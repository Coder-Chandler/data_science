```python
# 基本信息
import numpy as np
import pandas as pd
from pandas import Series, DataFrame

# 股票数据的读取
import pandas_datareader as pdr

# 可视化
import matplotlib.pyplot as plt
import seaborn as sns

# time
from datetime import datetime
```

```python
start = datetime(2014, 9, 20)
alibaba = pdr.get_data_yahoo('BABA', start=start)
amazon = pdr.get_data_yahoo('AMZN', start=start)
```

```python
alibaba.head()
```

```python
amazon.head()
```

```python
alibaba['Adj Close'].plot(legend=True)
plt.show()
```

```python
alibaba['Adj Close'].plot(legend=True)
amazon['Adj Close'].plot(legend=True)
plt.show()
```

```python
alibaba.head()
```

```python
alibaba['high-low'] = alibaba['High'] - alibaba['Low']
```

```python
alibaba.head()
```

```python
alibaba['high-low'].plot(legend=True)
plt.show()
```

```python
alibaba['daily-retuen'] = alibaba['Adj Close'].pct_change()
```

```python
alibaba.sort_values('daily-retuen',ascending=False).head()
```

```python
alibaba['daily-retuen'].plot(legend=True, figsize=(18,10), linestyle='--', marker='o')
plt.show()
```

```python
sns.distplot(alibaba['daily-retuen'].dropna(), bins=100, color='purple')
plt.show()
```
