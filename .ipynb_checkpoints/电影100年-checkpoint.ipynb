# 电影具体发行日期

下面尝试以具体的电影发布的年月日来分析，之前已经构建了年度，现在提取具体日期。

```python
month_order = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
day_order = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
```

```python
# 提取月份函数
def get_month(x):
    try:
        return month_order[int(str(x).split('-')[1]) - 1]
    except:
        return np.nan
    
# 提取天的函数
def get_day(x):
    try:
        year, month, day = (int(i) for i in x.split('-'))    
        answer = datetime.date(year, month, day).weekday()
        return day_order[answer]
    except:
        return np.nan
```

```python
df['day'] = df['release_date'].apply(get_day)
df['month'] = df['release_date'].apply(get_month)
```

```python
df[['title','year','month','day']].head()
```

```python
plt.figure(figsize=(12,6))
plt.title("Number of Movies released in a particular month from 1899 to 2017")
sns.countplot(x='month', data=df, order=month_order)
```

每年一月份电影发行最多，也就是说新年开始的时间是新片发行最多的时候！大家以后元旦期间多关注电影，会有很多大片。

那么哪些月份的电影挣钱最多呢，考虑超过超过1亿美金的电影。

```python
month_mean = pd.DataFrame(df[df['revenue'] > 100000000].groupby('month')['revenue'].mean())
month_mean['mon'] = month_mean.index
month_mean.head()
```

```python
plt.figure(figsize=(12,6))
plt.title("Average Gross by the Month for Blockbuster Movies")
sns.barplot(x='mon', y='revenue', data=month_mean, order=month_order)
```

```python
# 说明：coor() 相关系数
# 皮尔逊积矩相关系数（英语：Pearson product-moment correlation coefficient，又称作 PPMCC或PCCs[1], 常用r或Pearson's r表示）
# 用于度量两个变量X和Y之间的相关（线性相关），其值介于-1与1之间。在自然科学领域中，该系数广泛用于度量两个变量之间的相关程度。
# 它是由卡尔·皮尔逊从弗朗西斯·高尔顿在19世纪80年代提出的一个相似却又稍有不同的想法演变而来。这个相关系数也称作“皮尔森相关系数r”。
sns.set(font_scale=1)
corr = df.corr()
corr.head()
```

```python
# 创建元素全为 0 的数组, 类似 np.ones
mask = np.zeros_like(corr)
mask
```

```python
mask[np.triu_indices_from(mask)] = True
mask
```

```python
with sns.axes_style("white"):
    plt.figure(figsize=(9,9))
    ax = sns.heatmap(corr, mask=mask, vmax=.3, square=True, annot=True)
```

图中系数越高，说明相关性也就越大，比如说预算和收入最高，这张图也是对之前的相关性分析的一个汇总。

```python
# 特征工程
def feature_engineering(df):
    # belongs_to_collection变成一个布尔变量。 1表示电影是集合的一部分，而0表示不是
    df['belongs_to_collection'] = df['belongs_to_collection'].apply(lambda x: 0 if x == np.nan else 1)
    
    # genres转换成流派的数量
    for genre in genres_train:
        df[str(genre)] = df['genres'].apply(lambda x: 1 if genre in x else 0)
    df['genres'] = df['genres'].apply(lambda x: len(x))
    
    # homepage将被转换为一个布尔变量，将指示电影是否有主页
    df['homepage'] = df['homepage'].apply(lambda x: 0 if x == np.nan else 1)
    
    # original_language将被一个名为is_foreign的功能所取代，以表示某部电影是英语还是其他语言
    df['is_english'] = df['original_language'].apply(lambda x: 1 if x=='en' else 0)
    df = df.drop('original_language', axis=1)
    
    # production_companies将被替换为合作制作电影的制作公司的数量
    df['production_companies'] = df['production_companies'].apply(lambda x: len(x))
    
    # production_countries将被电影拍摄国家的数量所取代
    df['production_countries'] = df['production_countries'].apply(lambda x: len(x))
    
    # day将被转换为二进制功能，以表明电影是否在星期五发行
    df['is_Friday'] = df['day'].apply(lambda x: 1 if x=='Fri' else 0)
    df = df.drop('day', axis=1)
    
    # month将被转换成一个变量，表明month是否为假期
    df['is_Holiday'] = df['month'].apply(lambda x: 1 if x in ['Apr', 'May', 'Jun', 'Nov'] else 0)
    df = df.drop('month', axis=1)
    
    # 丢掉'title', 'cast', 'director'
    df = df.drop(['title', 'cast', 'director'], axis=1)
    
    return df
```

```python
def classification_engineering(df):
    
    # genres转换成流派的数量
    for genre in genres_train:
        df['is_' + str(genre)] = df['genres'].apply(lambda x: 1 if genre in x else 0)
    df['genres'] = df['genres'].apply(lambda x: len(x))
    
    # 丢弃homepage
    df = df.drop('homepage', axis=1)
    
    # 如果是英语片就为1，不是就为0
    df['is_english'] = df['original_language'].apply(lambda x: 1 if x=='en' else 0)
    
    # 丢弃电影语言列
    df = df.drop('original_language', axis=1)
    
    # 制作公司改为制作公司数量
    df['production_companies'] = df['production_companies'].apply(lambda x: len(x))
    
    # 拍摄地改为拍摄地数量
    df['production_countries'] = df['production_countries'].apply(lambda x: len(x))
    
    # day改为是否是周五发行，是就为1，不是就为0
    df['is_Friday'] = df['day'].apply(lambda x: 1 if x=='Fri' else 0)
    
    # 丢掉day这一列
    df = df.drop('day', axis=1)
    
    # 如果电影是在4、5、6、11月发行就为1，否则为0
    df['is_Holiday'] = df['month'].apply(lambda x: 1 if x in ['Apr', 'May', 'Jun', 'Nov'] else 0)
    
    # 丢弃month列
    df = df.drop('month', axis=1)
    
    # 丢弃'title', 'cast', 'director'列
    df = df.drop(['title', 'cast', 'director'], axis=1)
    
    df['runtime'] = df['runtime'].fillna(df['runtime'].mean())
    df['vote_average'] = df['vote_average'].fillna(df['vote_average'].mean())
    df = df.drop('crew', axis=1)
    return df
    
```

简单的推荐
根据电影的受欢迎程度和流派为每个用户提供一般推荐。这个推荐者背后的基本思想是，更受欢迎，更受好评的电影将有更高的被普通观众喜欢的可能性。这个模型并没有给出基于用户的个性化推荐。

也就是说vote_count投票数必须大于等于434票才能放进这张表中

```python
# 按照IMDB的评分公式计算
def weighted_rating(x):
    v = x['vote_count']
    R = x['vote_average']
    return (v*R/(v+m)) + (m*C/(m+v))

# 把计算的wr结果添加至上面的电影表中
qualified['wr'] = qualified.apply(weighted_rating, axis=1)

qualified = qualified.sort_values('wr', ascending=False).head(250)
```

```python
s = md.apply(lambda x: pd.Series(x['genres']),axis=1).stack().reset_index(level=1, drop=True)
s.name = 'genre'
gen_md = md.drop('genres', axis=1).join(s)
```

```python
gen_md.head()
```

```python
# 建立电影类型图表，分位数从95放宽到85
def build_chart(genre, percentile=0.85):
    
    # 调用函数传入指定类型，得出指定类型电影的推荐表格
    df = gen_md[gen_md['genre'] == genre]
    
    # 过滤掉票数为空的数据
    vote_counts = df[df['vote_count'].notnull()]['vote_count'].astype('int')
    
    # 过滤掉平均得票为空的数据
    vote_averages = df[df['vote_average'].notnull()]['vote_average'].astype('int')
    
    # 求得IMDB公式中的C
    C = vote_averages.mean()
    
    # 放宽分位数至85
    m = vote_counts.quantile(percentile)
    
    
    # 过滤出满足条件的电影
    qualified = df[(df['vote_count'] >= m) & (df['vote_count'].notnull()) & (
        df['vote_average'].notnull())][['title', 'year', 'vote_count', 'vote_average', 'popularity']]
    
    # 过滤出拍摄国家名称
    qualified['country'] = df['production_countries'].fillna('[]').apply(
    literal_eval).apply(lambda x: [i['name'] for i in x] if isinstance(x, list) else [])
    
    # 数据类型转换
    qualified['vote_count'] = qualified['vote_count'].astype('int')
    qualified['vote_average'] = qualified['vote_average'].astype('int')
    
    # 按照IMDB公式计算分值
    qualified['wr'] = qualified.apply(
        lambda x: (x['vote_count']/(x['vote_count']+m) * x['vote_average']) + (m/(m+x['vote_count']) * C), axis=1)
    qualified = qualified.sort_values('wr', ascending=False).head(250)
    
    return qualified
```
