```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from pandas import Series, DataFrame 
```

```python
x = np.linspace(0.0, 5.0)
y1 = np.sin(np.pi*x)
y2 = np.sin(np.pi*x*2)
```

```python
plt.subplot(2,2,1)
plt.plot(x, y1, 'b--')
plt.ylabel('y1')
plt.subplot(2,2,2)
plt.plot(x,y2, 'r--')
plt.ylabel('y2')
plt.xlabel('x')
plt.subplot(2,2,3)
plt.plot(x,y1, 'r*')
plt.ylabel('y3')
plt.subplot(2,2,4)
plt.plot(x,y1, 'b*')
plt.ylabel('y4')
```

```python
plt.show()
```

```python
figure, ax = plt.subplots(2,2)
ax[0][0].plot(x, y1, 'r')
ax[0][1].plot(x, y2, 'b')
ax[1][0].plot(x, y2, 'b--')
ax[1][1].plot(x, y1, 'r--')
plt.show()
```
