```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from pandas import Series, DataFrame 
```

```python
s1 = Series(np.random.randn(1000)).cumsum()
s2 = Series(np.random.randn(1000)).cumsum()
```

```python
s1.plot(grid=True, label='S1', 
        title='This is Series', style='--')
s2.plot(label='S2')
plt.legend()
```

```python
plt.show()
```

```python
fig, ax = plt.subplots(2,1)
```

```python
ax[0].plot(s1)
ax[1].plot(s2)
```

```python
plt.show()
```

```python
fig, ax = plt.subplots(2,1)
```

```python
s1[0:20].plot(ax=ax[0], kind='bar',label='S1', title='this is Series', grid=True)
s2[20:40].plot(ax=ax[1], label='S2', grid=True, style='r')
```
