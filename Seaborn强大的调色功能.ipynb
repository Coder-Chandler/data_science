```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
from pandas import Series, DataFrame 
```

```python
def sinplot():
    x = np.linspace(0, 14, 100)
    plt.figure(figsize=(8,6))
    for i in range(4):
        plt.plot(x, np.sin(x+i)*(i+0.75), label='sin(x+%s)*(%s+0.75)'%(i,i))
    plt.legend()
    plt.show()
```

```python
sinplot()
```
